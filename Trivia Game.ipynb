```python
class Stack:
    def __init__(self):
        """atribiutes"""
        self.stack=[]
        self.n = 0
        self.top = None
    def add(self,item):
        """
        Add items to stack
        
        """
        self.stack.append(item)
        self.n +=1
        self.top= len(self.stack) -1
    def pop(self):
        """return top item/last item placed"""
        if self.n==0:
            return 'The Stack is empty'
        else:
            self.result=self.stack[self.top]
            self.stack=self.stack[:-1]
            self.n-=1
            self.top = self.n - 1
            return self.result
    def is_empty(self):
        """Check if stack is empty"""
        if self.n==0:
            return True
        else:
            return False
    def size(self):
        """
        return the size of the stack
        """
        return self.n
    
    def peak(self):
        """
        peak at next item to be removed from stack
        """
        return self.stack[self.top]
```

```python
class Queue:
    def __init__(self):
        self.queue=[]
        self.n = 0
        self.bottom = 0
    def add(self, item):
        """
        add an element to queue
        """
        self.queue.append(item)
        self.n +=1
    def pop(self,item):
        """
        return top item/first item placed
        """
        if self.n==0:
            return 'The Queue is empty'
        else:
            self.result = self.queue[self.bottom]
            self.queue = self.queue[1:]
            self.n-=1
    def is_empty(self):
        """Check if the Queue is empty"""
        if self.n==0:
            return True
        else:
            return False
    def size(self):
        """check how many elements remaining in queue"""
        return self.n
    def peak(self):
        """check the next item to be removed"""
        return self.queue[self.top]
```
